# CNOT gate

### Import section

```python
from qiskit import *
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit_aer import StatevectorSimulator, QasmSimulator, UnitarySimulator
import numpy as np
from math import pi

%config InlineBackend.figure_format = 'svg' # Makes the images look nice
```

### Create  circuit

```python
n = 2
n_q = n
n_b = n
```

### Bell $|\Phi^+\rangle = \frac{|00\rangle +|11\rangle}{\sqrt{2}}$

```python
qc = QuantumCircuit(n_q,n_b)

qc.h(0)
qc.cx(0,1)
#qc.h(0)
#qc.h(1)
qc.measure(0,0)
qc.measure(1,1)

display(qc.draw(output='mpl'))
#backend = StatevectorSimulator()
#job = backend.run(qc).result()
#display(job.get_statevector())
```

```python
backend = QasmSimulator()
counts = backend.run(qc,shots = 1000).result().get_counts()
plot_histogram(counts)
```

### Bell $|\Phi^-\rangle = \frac{|00\rangle -|11\rangle}{\sqrt{2}}$

```python
qc = QuantumCircuit(n_q,n_b)

qc.x(0)
qc.h(0)
qc.cx(0,1)
#qc.h(0)
#qc.h(1)
qc.measure(0,0)
qc.measure(1,1)

display(qc.draw(output='mpl'))

backend = StatevectorSimulator()
job = backend.run(qc).result()
display(job.get_statevector())
```

```python
# show that by increasing Nshots you converge to the expected result
Nshots = 1024*8
backend = QasmSimulator()
counts = backend.run(qc,shots = Nshots).result().get_counts()
plot_histogram(counts)
```

#### CNOT: $|+-\rangle \rightarrow |--\rangle$ 

```python
qc = QuantumCircuit(2,2)

# init in 01
#qc.x(1)

# CX con scambiati control e target
qc.h(0)
qc.h(1)
qc.cx(0,1)
qc.h(0)
qc.h(1)

qc.draw('mpl')

qc2 = QuantumCircuit(2,2)
qc2.cx(1,0)

unitary_simulator = UnitarySimulator() 
final_state = unitary_simulator.run(qc).result().get_unitary()-unitary_simulator.run(qc2).result().get_unitary()
np.round(np.real(final_state),2)
```

## CNOT identities

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi

qreg_q = QuantumRegister(2, 'q')
qc = QuantumCircuit(qreg_q)

q0=0
q1=1
parameter = np.pi/3

qc.cx(q0, q1)
qc.rx(parameter/2, q0)
qc.h(q0)
qc.rz(parameter/2, q1)
qc.cx(q0, q1) 
qc.h(q0)
qc.rz(-parameter/2, q1)
qc.cx(0,1)
qc.rx(-np.pi/2, q0)
qc.rx(np.pi/2, q1)

display(qc.draw('mpl'))

unitary_simulator = UnitarySimulator() 
U = unitary_simulator.run(qc).result().get_unitary()*np.exp(-1j*np.arctan(0.43/0.75))

print(np.round(np.real(U),2))
print(np.round(np.imag(U),2))

print(np.round(U,2))

qc.draw()
```

```python
qc = QuantumCircuit(n_q)

qc.h(0)
qc.h(1)
qc.cx(0,1)
qc.h(0)
qc.h(1)

display(qc.draw('mpl'))

unitary_simulator = UnitarySimulator() 
U = unitary_simulator.run(qc).result().get_unitary()

print(np.round(np.real(U),2))
```

## SWAP gate

### Decomposition in CNOTs

```python
qc = QuantumCircuit(2)

qc.cx(0,1)
qc.cx(1,0)
qc.cx(0,1)

display(qc.draw('mpl'))

unitary_simulator = UnitarySimulator() 
U = unitary_simulator.run(qc).result().get_unitary()

print(np.real(U))
```

### This other circuit produces the same effect (SWAP is symmetric)

```python
qc = QuantumCircuit(2)

qc.cx(1,0)
qc.cx(0,1)
qc.cx(1,0)

display(qc.draw('mpl'))

unitary_simulator = UnitarySimulator() 
U = unitary_simulator.run(qc).result().get_unitary()

print(np.real(U))
```

### Decomposition in CNOTs

```python
qc = QuantumCircuit(3)

qc.h(0)
qc.cx(1,0)
qc.tdg(0)
qc.cx(2,0)
qc.t(0)
qc.cx(1,0)
qc.tdg(0)
qc.cx(2,0)
qc.t(0)
qc.t(1)
qc.cx(2,1)
qc.h(0)
qc.t(2)
qc.tdg(1)
qc.cx(2,1)

display(qc.draw('mpl'))

unitary_simulator = UnitarySimulator() 
U = unitary_simulator.run(qc).result().get_unitary()

print(np.round(np.real(U),2))
```
